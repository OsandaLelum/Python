```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset from Seaborn
tips = sns.load_dataset('tips')

# Plot a bar chart of the total bill amount
sns.barplot(data=tips, x='day', y='total_bill', ci=None)

# Add labels to the plot
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')
plt.title('Total Bill Amount by Day of the Week')

# Display the plot
plt.show()
```

```python
sns.boxplot(x="total_bill", data=tips)

# Show the plot
plt.show()
```

```python
#Basic Heat Map
import seaborn as sns
import numpy as np

# Create random data
data = np.random.rand(10, 10)

# Plot heat map
sns.heatmap(data)
```

```python
#heat mapwith anotations
import seaborn as sns
import numpy as np

# Create random data
data = np.random.rand(10, 10)

# Plot heat map with annotations
sns.heatmap(data, annot=True)
```

```python
#heat map with color map
import seaborn as sns
import numpy as np

# Create random data
data = np.random.rand(10, 10)

# Define color map
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Plot heat map with color map
sns.heatmap(data, cmap=cmap)
```

```python
#correlation matrix with heat map

# Create random data
data = np.random.rand(10, 10)

# Compute correlation matrix
corr = np.corrcoef(data)

# Plot correlation matrix heat map
sns.heatmap(corr)
```

```python
plt.figure(figsize=(4,3))
sns.pairplot(data=iris, hue='species', markers=['o','s','D'])
plt.title('Pair plot')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.violinplot(x="day", y="total_bill",data=tips)
plt.title('Violin plot')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.swarmplot(data=tips,x="day", y="total_bill",hue='sex')
plt.title('Swarmplot')
plt.show()
```
