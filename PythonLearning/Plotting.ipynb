```python
import numpy as np 

import matplotlib.pyplot as plt
```

```python
x=np.arange(-3.14,3.14,0.01)
ySin = np.sin(x)
yCos = np.cos(x)
plt.figure(figsize=(5,5))
plt.subplot(2,1,1)
plt.plot(x,ySin,label="sin(x)")
plt.plot(x,yCos,label="cos(X)")
plt.xlabel('x')
plt.legend()
plt.grid()
plt.title('Graph Name')

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 100 * np.random.rand(100)

# Create a scatter plot with color mapping
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')

# Customize the plot
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Color Mapping')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = ['A', 'B', 'C', 'D']
y = [3, 4, 5, 6]
error = [0.5, 0.2, 0.3, 0.8]

# Create a bar chart with error bars
plt.bar(x, y, yerr=error, align='center', alpha=0.7)

# Customize the plot
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Error Bars')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
data1 = np.random.randn(1000)
data2 = np.random.randn(500)

# Create a histogram with multiple data sets
plt.hist([data1, data2], bins=30, color=['b', 'r'], label=['Data 1', 'Data 2'])

# Customize the plot
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Multiple Data Sets')
plt.legend()

# Display the plot
plt.show()
```
