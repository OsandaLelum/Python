K-means clustering is an iterative algorithm used to partition a dataset into K distinct clusters. It aims to minimize the within-cluster sum of squares (WCSS), also known as inertia. Here is an overview of the K-means clustering algorithm:

kmeans.png

1. Initialization:
   - Select the number of clusters, K.
   - Initialize K centroids randomly or using a specific initialization method (e.g., k-means++).
   - Assign each data point to its nearest centroid.

2. Iterative Assignment and Update:
   - Repeat the following steps until convergence:
     - For each data point, calculate the distance to each centroid and assign it to the nearest centroid.
     - Update the centroids by calculating the mean of all data points assigned to each centroid.

3. Convergence:
   - The algorithm converges when the centroids no longer move significantly or when a predefined number of iterations is reached.

4. Final Clustering:
   - The final clusters are obtained when the algorithm converges.
   - Each data point belongs to the cluster represented by its nearest centroid.

5. Evaluation and Refinement:
   - Assess the quality of the clustering results using evaluation metrics (e.g., silhouette score, WCSS).
   - Adjust the number of clusters, K, if necessary, and rerun the algorithm.

K-means clustering has some limitations:
- It is sensitive to the initial centroid positions, and different initializations can lead to different clustering results.
- It assumes clusters are spherical and equally sized, which may not hold for all types of data.
- It can be affected by outliers that can significantly influence the position of centroids.

To improve the performance and address some limitations, variations of K-means clustering have been developed, such as:
- K-means++ initialization: A seeding method that improves the initial centroid selection, leading to more accurate and stable results.
- Mini-Batch K-means: A variant that uses mini-batches of data for faster convergence in large datasets.
- Hierarchical K-means: A hierarchical version that builds a hierarchy of clusters instead of a single partitioning.

When using K-means clustering, it is essential to preprocess the data, select an appropriate value for K, and evaluate the results to ensure they align with your objectives.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
X,y = make_blobs(n_samples=10000,centers=3,n_features=2,random_state=23)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_train = train_test_split(X,y,test_size=0.333,random_state=42)
print(X_train.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []  # Create an empty list to store the WCSS values

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++')  # Create a K-means object with k clusters
    kmeans.fit(X_train)  # Fit the K-means model to the training data
    wcss.append(kmeans.inertia_)  # Append the WCSS value to the list

```
